* We have already regularised a fully grown Decsion Tree and arrived at much efficient Regularised Decision tree model. Though the metrics are encouraging there is still scope for improvement. Hence we are going for Random Forest (a speacial bagging method).

* Decision Tree is greedy model. It overfits the model. To overcome it, we have Random FOrest technique. 

* Unlike decision tree, in random forest bagging(bootstrap sampling) has been used. And also that feature selection is done here, as one sqrt(n) features has been used for each bag. Hence we go for Random FOrest

```python
from sklearn.ensemble import RandomForestClassifier
```

* we use customised grid search as Built in grid search doesn't care about Variance Error

```python
rf_ve = []
rf_be = []
for n in np.arange(1,25):
    RF = RandomForestClassifier(n_estimators= n, criterion= 'entropy', random_state= 0)
    kf = KFold(shuffle= True, n_splits= 5, random_state= 0)
    roc = cross_val_score(RF, xtrain, ytrain, cv=kf, scoring='roc_auc')
    rf_be.append(np.mean(1-roc))
    rf_ve.append(np.std(roc, ddof=1))
```

```python
x_ax = np.arange(len(rf_ve))
plt.plot(x_ax, rf_ve)
plt.show()
```

```python
np.min(rf_ve), np.argmin(rf_ve), rf_be[2]
```

```python
# Regularised RF model

# index starts with 0 in list(rf_be) but corresponding n_estimator is 1. Hence, best n_estimator is 2+1 = 3

rf = RandomForestClassifier(n_estimators= 3, criterion= 'entropy', random_state = 0)
```

```python
kf = KFold(shuffle= True, n_splits= 5, random_state= 0)
roc = cross_val_score(rf, xtrain, ytrain, cv=kf, scoring='roc_auc')
print('Bias Error :', np.mean(1-roc))
print('Variance Error :',np.std(1-roc, ddof=1))
```

* It is very evident that on using the RandomForest classifier (Bagging ALgorithm) our variance error got reduced by larger extent, but bias error got increased by fractions which is not in our control.

* We need to compare mall the models we have built to conclude which is the best suited for this dataset.

```python
# Summarize all the models

dt_model = DecisionTreeClassifier(random_state = 0)
dt_reg = DecisionTreeClassifier(criterion= 'entropy', max_depth=6, random_state = 0)
rf = RandomForestClassifier(n_estimators= 3, criterion= 'entropy', random_state = 0)
```

```python
models = []

models.append(('DecisionTree', dt_model))
models.append(('DT_regularised', dt_reg))
models.append(('RF', rf))
```

```python
models
```

```python
results = []
names = []
for name, model in models:
    kf = KFold(shuffle= True, n_splits= 5, random_state= 0)
    roc = cross_val_score(model, xtrain, ytrain, cv=kf, scoring='roc_auc')
    results.append(roc)
    names.append(name)
    print("%s: %f (%f)" % (name, np.mean(1-roc), np.std(roc, ddof=1)))

fig = plt.figure()
fig.suptitle('Algorithm Comparison')
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show()
```

* Decision Tree is the worst performing model out of the three. Lets comapre the remaining two.

```python
results = []
names = []
for name, model in models[1:]:
    kf = KFold(shuffle= True, n_splits= 5, random_state= 0)
    roc = cross_val_score(model, xtrain, ytrain, cv=kf, scoring='roc_auc')
    results.append(roc)
    names.append(name)
    print("%s: %f (%f)" % (name, np.mean(1-roc), np.std(roc, ddof=1)))

fig = plt.figure()
fig.suptitle('Algorithm Comparison')
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show()
```
